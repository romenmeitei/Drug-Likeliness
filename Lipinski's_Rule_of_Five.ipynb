```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from rdkit import Chem
from rdkit.Chem import Descriptors

# ðŸ”¹ Load compound dataset (Assumes CSV file with 'SMILES' column)
from google.colab import files
uploaded = files.upload()  # Upload a CSV file

df = pd.read_csv(list(uploaded.keys())[0])  # Load uploaded file
if 'SMILES' not in df.columns:
    raise ValueError("CSV file must contain a 'SMILES' column!")

# ðŸ”¹ Define Lipinskiâ€™s Rule of Five function
def lipinski_rules(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return None, None  # Invalid SMILES

    mw = Descriptors.MolWt(mol)         # Molecular Weight
    logp = Descriptors.MolLogP(mol)     # LogP (Lipophilicity)
    hbd = Descriptors.NumHDonors(mol)   # Hydrogen Bond Donors
    hba = Descriptors.NumHAcceptors(mol)  # Hydrogen Bond Acceptors

    # Lipinski Violations
    violations = sum([
        mw > 500,    # MW > 500 Da
        logp > 5,    # LogP > 5
        hbd > 5,     # HBD > 5
        hba > 10     # HBA > 10
    ])

    return violations, mw, logp, hbd, hba

# ðŸ”¹ Apply function to all compounds
df[['Violations', 'MW', 'LogP', 'HBD', 'HBA']] = df['SMILES'].apply(lambda x: pd.Series(lipinski_rules(x)))

# ðŸ”¹ Summary statistics
total_compounds = df.shape[0]
drug_like_count = df[df['Violations'] == 0].shape[0]
percent_drug_like = (drug_like_count / total_compounds) * 100

print(f"ðŸ“Š Total Compounds: {total_compounds}")
print(f"âœ… Drug-Like Compounds: {drug_like_count} ({percent_drug_like:.2f}%)")
print(f"ðŸš¨ Compounds with Lipinski Violations: {total_compounds - drug_like_count}")

# ðŸ”¹ Plot violations
plt.figure(figsize=(6,4))
df['Violations'].value_counts().sort_index().plot(kind='bar', color=['green', 'orange', 'red'])
plt.xlabel("Number of Lipinski Violations")
plt.ylabel("Number of Compounds")
plt.title("Lipinski's Rule of Five Violations of the Phytocompounds present in Terminilia citrina")
plt.xticks(rotation=0)
plt.show()

# ðŸ”¹ Save results
df.to_csv("Lipinski_Results.csv", index=False)
print("ðŸ“‚ Results saved as 'Lipinski_Results.csv'. Download it from Colab.")
```
